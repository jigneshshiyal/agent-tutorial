```python
from langchain_community.llms import Ollama
from langchain.callbacks.streaming_stdout import StreamingStdOutCallbackHandler

# Streaming handler prints tokens as they are generated
streaming_handler = StreamingStdOutCallbackHandler()

# DO NOT include 'streaming=True' â€“ just pass the callback
llm = Ollama(
    model="gemma3n",
    callbacks=[streaming_handler],
)
```

```python
for i in llm.stream("hello"):
    print(i, end="", flush=True)
```

```python
from typing import TypedDict
# Define state
class EchoState(TypedDict):
    input: str
    output: str

# Echo node
def echo_node(state: EchoState) -> EchoState:
    prompt = f"You said: {state['input']}"
    response = llm(prompt)  # streaming happens automatically via callback
    return {"input": state["input"], "output": response}
```

```python
from langgraph.graph import StateGraph
# Build LangGraph
builder = StateGraph(EchoState)
builder.add_node("echo", llm)
builder.set_entry_point("echo")
builder.set_finish_point("echo")
graph = builder.compile()
```

```python
user_input = input("You: ")
for i in graph.stream(
    {"input": user_input, "output": ""},
    stream_mode="messages"
):
    print(f"Bot: {i['output']}", end="##", flush=True)
```

```python
user_input = input("You: ")
result = graph.invoke({"input": user_input})
print("\nFinal Output:", result["output"])
```
