```python
# retrievers
from typing import List
from langchain_core.documents import Document
from langchain_core.runnables import chain

@chain
def retriever(query: str) -> List[Document]:
    """Retrieve documents based on the query."""
    return vector_store.similarity_search(query)


retriever.invoke("What is the history of Nike?")
```

#### Using tool calling

```python
from pydantic import BaseModel, Field

class ResponseFormatter(BaseModel):
    answer: str = Field(description="The answer to the question.")
    followup_question: str = Field(description="A followup question the user could ask")
```

```python
model_with_tools = model.bind_tools([ResponseFormatter])

ai_msg = model_with_tools.invoke("What is capital of France?")
```

```python
# answer message
ai_msg.tool_calls[0]["args"]

# validate the response
pydantic_object = ResponseFormatter.model_validate(ai_msg.tool_calls[0]["args"])
print(pydantic_object.answer)
print(pydantic_object.followup_question)
```

### Extraction Chain

```python
# define schema for extraction
from pydantic import BaseModel, Field
from typing import Optional

class ExtractionSchema(BaseModel):
    name: Optional[str] = Field(description="The name of the person.")
    hair_color: Optional[str] = Field(description="The hair color of the person.")
    height_in_meters: Optional[float] = Field(description="The height of the person in meters.")

class MainExtractionSchema(BaseModel):
    extraction: List[ExtractionSchema] = Field(description="The extracted information from the text.")

model = ChatGoogleGenerativeAI(model="models/gemini-2.5-flash", api_key=GEMINI_API_KEY)
model = model.with_structured_output(schema=MainExtractionSchema)

from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder

prompt_template = ChatPromptTemplate.from_messages([
    (
        "system",
        "You are an expert extraction algorithm. "
        "Only extract relevant information from the text. "
        "If you do not know the value of an attribute asked to extract, "
        "return null for the attribute's value.",
    ),
    # Please see the how-to about improving performance with
    # reference examples.
    # MessagesPlaceholder('examples'),
    ("human", "{text}"),
])

text = "Alan Smith is 6 feet tall and has blond hair."
prompt = prompt_template.invoke({"text": text})

model.invoke(prompt)
```

### Examples

```python
from langchain_core.utils.function_calling import tool_example_to_messages

examples = [
    (
        "The ocean is vast and blue. It's more than 20,000 feet deep.",
        MainExtractionSchema(extraction=[])
    ),
    (
        "Alan Smith is 6 feet tall and has blond hair.",
        MainExtractionSchema(extraction=[ExtractionSchema(name='Alan Smith', hair_color='blond', height_in_meters=1.8288)])
    ),
]

messages = []

for txt, tool_call in examples:
    if tool_call.extraction:
        # This final message is optional for some providers
        ai_response = "Detected people."
    else:
        ai_response = "Detected no people."
    messages.extend(tool_example_to_messages(txt, [tool_call], ai_response=ai_response))
```
